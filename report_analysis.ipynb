```python
import pandas as pd
import matplotlib.pyplot as plt
import os

path='./csv/'
data_frame=pd.read_csv(path+'sales2019_1.csv')
data_frame.head(10)
```

```python
frames=[]
all_length=[]
for file in os.listdir(path):
    if file.endswith('.csv'):
        file_path=path + file
        data_frame1=pd.read_csv(file_path)
        frames.append(data_frame1)
        result=pd.concat(frames)
        length_1month=len(data_frame1.index)
        all_length.append(length_1month)

df=result
df.to_csv('./result/result_report.csv',index=False)
print(result)
```

```python
df['Tháng']=df['Order Date'].str[0:2]
df.head()
```

```python
print(set(df['Tháng']))
```

```python
df=df.dropna(how="all")
df=df[df['Tháng']!='Or']
df.head()
```

```python
print(df['Quantity Ordered'].dtypes,df['Price Each'].dtypes)
```

Task 1:Report

```python
df['Doanh thu tháng']= pd.to_numeric(df['Quantity Ordered'],downcast='float')* pd.to_numeric(df['Price Each'],downcast='float')
df.head()
```

```python
df=df.rename({'Order ID':'Mã đặt','Product':'Sản phẩm','Quantity Ordered':'Số lượng','Price Each':'Giá bán','Order Date':'Ngày Đặt','Purchase Address':'Địa chỉ'},axis=1)
df.to_csv('./result/vi_report.csv',encoding='UTF-8',header=True)
df.head()
```

```python
moving_col=df.pop('Doanh thu tháng')
df.insert(4,'Doanh thu tháng',moving_col)
df.head()
```

```python
sales_value=df.groupby('Tháng').sum()['Doanh thu tháng']
sales_value.max()
```

```python
months=range(1,13)
plt.bar(x=months,height=sales_value)
plt.xticks(months)
plt.xlabel('Tháng')
plt.ylabel('Doanh thu')
plt.show()
```

```python
address_to_city=lambda address:address.split(",")[1].strip()
```

```python
df['Thành phố']=df['Địa chỉ'].apply(address_to_city)
df.head()
```

```python
values_city= df.groupby('Thành phố').sum()['Doanh thu tháng']
print(values_city)
```

```python
cities=[]
for city,sales in values_city.items():
    cities.append(city)
print(cities)

plt.bar(x=cities,height=values_city)
plt.xticks(cities,rotation=65)
plt.xlabel('Thành phố')
plt.ylabel('Doanh thu')
plt.show()
```
